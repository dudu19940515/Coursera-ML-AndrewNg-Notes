```python
import pandas as pd
data_2 = pd.read_csv('ex1data2.txt',delimiter=',',names = ['square','bedrooms','price'])
data_2.head()
```

## 标准化数据

```python
def feature_normalize(df):
    return df.apply(lambda x: (x-x.mean())/x.std())
```

```python
data_imp = feature_normalize(data_2)
data_imp.head()
#归一化=后的数据
```

## 计算Cost Function大小

```python
data_imp.columns.size
```

```python
def get_x(df):
    data = pd.DataFrame({'ones':np.ones(len(df))})
    df = pd.concat((data,df),axis =1)
    return df.iloc[:,:-1].as_matrix()
```

```python
x = get_x(data_imp)
# 形成x的矩阵\
x.shape
```

```python
def get_y(df):
    return df.iloc[:,-1].as_matrix()
```

```python
y = get_y(data_imp)
y = y.reshape((-1,1))
y.shape
```

```python
def computeCostMulti(theta,x,y):
    num = x.shape[0]
    feature_num = x.shape[1]
    inner = x@theta - y
    J = inner.T@inner
    return J/(2*num)
```

## 计算梯度下降

```python
def gradient_Descent(theta,x,y,alpha = 0.01,epoch = 500):
    m = x.shape[0]
    J = np.zeros([epoch+1,1])
    J[0,0] = computeCostMulti(theta,x,y)
    for i in range(epoch):
        theta =theta - alpha*x.T@(x@theta-y)/m
        J[i+1,0] = computeCostMulti(theta,x,y)
    return J,theta
        
```

```python

theta = np.zeros([x.shape[1],1])
alpha = 0.01
epoch = 500
J = np.zeros([epoch+1,1])
theta.shape
```

```python
J,theta = gradient_Descent(theta,x,y, alpha = alpha,epoch = epoch)
```

```python
computeCostMulti(theta,x,y)
```

```python
sns.tsplot(np.arange(epoch+1),J[:,0])
plt.xlabel('Cost')
plt.ylabel('No.iteration')
```

## learning Rate （学习率）

```python
alpha_candidate = [0.01,0.03,0.1,0.3]
epoch = 50
theta = np.zeros((3,1))
fig,ax = plt.subplots(figsize = (16,8))
for alpha in alpha_candidate:
    cost_data,_ = gradient_Descent(theta,x,y,alpha =alpha,epoch = epoch)
    ax.plot(np.arange(epoch+1),cost_data[:,0],label = alpha)
plt.xlabel('NO.iterations')  
plt.ylabel('cost_data')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
